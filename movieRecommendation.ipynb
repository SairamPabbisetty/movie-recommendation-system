```python
import numpy as np 
import pandas as pd
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity 

movies_data = pd.read_csv("movies.csv")

selected_features = ['title', 'genres', 'keywords', 'tagline', 'cast', 'director', 'release_date']

for feature in selected_features :
    movies_data[feature] = movies_data[feature].fillna('')
    
movies_data["release_date"] = [movies_data["release_date"][i][:4] for i in range(len(movies_data))]

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '+movies_data['release_date']

vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)

similarity = cosine_similarity(feature_vectors)

req_features = ['title', 'genres', 'tagline', 'director', 'release_date']
print(req_features)

feature_name = input("Select a feature name from the list: ")
if feature_name in req_features :
    if feature_name == 'release_date' :
        feature_value = input("Enter the "+ feature_name+"(Year): ")
    else :
        feature_value = input("Enter the "+ feature_name+": ")
        
else :
    print("Enter the feature present in the list!")

list_of_all_titles = movies_data[feature_name].tolist()

find_close_match = difflib.get_close_matches(feature_value, list_of_all_titles)
close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data[feature_name] == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similarity_score = sorted(similarity_score, key=lambda x:x[1], reverse=True)

print('Movies for you \n')

i = 1
for movie in sorted_similarity_score :
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if i<=10 :
        print(i, '.', title_from_index)
        i += 1
```
